```python
import pandas as pd
import glob
import os
import json, re
```

```python
#get all csv

all_data = []
path = "../meld_data"
csv_files = sorted(glob.glob(os.path.join(path, '*.csv')))
```

```python
#convert to df

for csv in csv_files:
    df = pd.read_csv(csv)
    all_data.append(df)

df_final = pd.concat(all_data, ignore_index = True)
```

```python
#sort the df

df_final['StartTime'] = pd.to_timedelta(df_final['StartTime'].str.replace(',', '.'))
df_final['EndTime'] = pd.to_timedelta(df_final['EndTime'].str.replace(',', '.'))

df_sorted = df_final.sort_values(by=['Season', 'Episode', 'StartTime']).reset_index(drop=True)
```

```python
#get the durations of all dialogue in s1-s6 from transcripts



transcript_rows = []
base_dir = "/home/cleode5a7/Desktop/analyses_friends_annotations/friends_annotations/annotation_results/Speech2Text/"

for season_folder in glob.glob(os.path.join(base_dir, "s[1-6]")):
    # All JSON files starting with 'friends'
    for file in glob.glob(os.path.join(season_folder, "friends*.json")):
        fname = os.path.basename(file)
        m = re.search(r"s(\d+)e(\d+)", fname)
        if m:
            season_num = int(m.group(1))
            ep_num = int(m.group(2))
        else:
            continue  # skip files that don't match

        with open(file) as f:
            data = json.load(f)

        words = data["results"]["channels"][0]["alternatives"][0]["words"]

        for w in words:
            transcript_rows.append({
                "Season": season_num,
                "Episode": ep_num,
                "Start": w["start"],
                "End": w["end"]
            })

# Build DataFrame
transcript_df = pd.DataFrame(transcript_rows)
transcript_df["Duration"] = transcript_df["End"] - transcript_df["Start"]

# Total spoken time per episode
episode_times = transcript_df.groupby(["Season", "Episode"])["Duration"].sum().reset_index()
print(episode_times)

season_times = transcript_df.groupby(["Season"])["Duration"].sum().reset_index()

print(season_times)
```

```python

def to_seconds(t):
    h, m, s = t.replace(',', '.').split(':')
    return int(h)*3600 + int(m)*60 + float(s)

# Load annotations
annotations = df_sorted

# Annotated time per season
# Annotated time per season
annotated_times = {
    season: sum(
        (e - s).total_seconds()
        for s, e in zip(
            annotations[annotations["Season"] == season]["StartTime"],
            annotations[annotations["Season"] == season]["EndTime"]
        )
    )
    for season in range(1, 7)
}
 # convert to df
annotated_df = pd.DataFrame(list(annotated_times.items()), columns=["Season", "AnnotatedDuration"])
```

```python
season_comparison = pd.merge(season_times, annotated_df, on="Season")

# Compute coverage %
season_comparison["CoveragePercent"] = (season_comparison["AnnotatedDuration"] / season_comparison["Duration"]) * 100

print(season_comparison)
```
